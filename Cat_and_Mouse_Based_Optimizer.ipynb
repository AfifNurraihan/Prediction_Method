#**Optimasi menggunakan CMBO**

## Import Library

```python
import math
import numpy as np
import pandas as pd
from random import randrange
```

## Inisialisasi Awal

```python
lb         = -100 #Batas Bawah
ub         = 100  #Batas Atas
dm         = 2    #Dimensi
population = 20   #Banyak Search Agent
max_iter   = 100
```

## Pembentukan Populasi Awal

```python
def generate_agent(lb, ub,size):
  agent = lb + np.random.rand(*size)*(ub-lb)
  return agent
```

```python
agent = generate_agent(lb,ub,size=(population,dm))
print('Populasi Awal : \n',agent)
```

## Menghitung Fungsi Objektif pada Populasi Awal

```python
def objective_function(agent):
  # x**2 + x**2
  F_obj = agent[:,0]**2 + agent[:,1]**2
  return F_obj
```

```python
F_obj = objective_function(agent)
print('Fungsi Objektif : \n',F_obj)
```

## Urutkan Populasi Awal berdasarkan Fungsi Objektif Terurut

```python
sorted_index = F_obj.argsort()
F_obj        = F_obj[sorted_index]
agent_ori    = agent.copy()
agent        = agent[sorted_index]
```

## Improve Pembagian Populasi Tikus dan Kucing

```python
# Improve populasi Tikus 50%, Kucing 50%
pres_mice  = 50                                # Dalam Persen
gen_mice   = round(population*(pres_mice/100)) # Banyaknya populasi tikus

upper_half = slice(gen_mice)
lower_half = slice(gen_mice, len(agent))
```

```python
#Populasi Tikus
mice_pop = agent[upper_half]
mice_obj = F_obj[upper_half]

#Populasi Kucing
cat_pop = agent[lower_half]
cat_obj = F_obj[lower_half]

print('Populasi Tikus  : \n', mice_pop,'\n')
print('Populasi Kucing : \n', cat_pop,'\n')
```

## Kucing Mengejar Tikus

```python
# Update pergerakan kucing dengan sebarang tikus indeks berbeda
def update_cat_movement(cat_pop, mice_pop):
  random_index = randrange(len(mice_pop))
  I = np.round(1 + np.random.rand(len(cat_pop), dm))
  r = np.random.rand()
  new_pop = cat_pop + r *(mice_pop[random_index] - I * cat_pop)
  return new_pop
```

```python
cat_pop_move = update_cat_movement(cat_pop, mice_pop)
cat_obj_move = objective_function(cat_pop_move)

print('Populasi Kucing Bergerak : \n',cat_pop_move)
print('\n')
print('Fungsi Objektif Populasi Kucing Bergerak : \n',cat_obj_move)
```

```python
lower_than = cat_obj_move < cat_obj
cat_obj[lower_than] = cat_obj_move[lower_than]
cat_pop[lower_than] = cat_pop_move[lower_than]
```

## Tikus bersembunyi ke dalam Sarang

```python
haven_pop = generate_agent(lb, ub,size=(mice_pop.shape))
haven_obj = objective_function(haven_pop)
print('Populasi Sarang : \n',haven_pop)
print('\n')
print('Fungsi Objektif Populasi Sarang : \n',haven_obj)
```

```python
def update_mice_escape(mice_pop, haven_pop, mice_obj, haven_obj):
  I = np.round(1 + np.random.rand(len(mice_pop), dm))
  r = np.random.rand()
  new_pop = mice_pop + r * (haven_pop - I * mice_pop) * np.sign(mice_obj - haven_obj)[:, np.newaxis]
  return new_pop
```

```python
mice_pop_esc = update_mice_escape(mice_pop, haven_pop, mice_obj, haven_obj)
mice_obj_esc = objective_function(mice_pop_esc)
print('Populasi Tikus Kabur : \n',mice_pop_esc)
print('\n')
print('Fungsi Objektif Populasi Tikus Kabur : \n',mice_obj_esc)          
```

```python
lower_than = mice_obj_esc < mice_obj
mice_obj[lower_than] = mice_obj_esc[lower_than]
mice_pop[lower_than] = mice_pop_esc[lower_than]
```

## Populasi Update

```python
agent = np.concatenate([mice_pop, cat_pop], axis=0)
print('Populasi Update : \n',agent)
```

```python
new_obj = np.concatenate([mice_obj, cat_obj], axis=0)
new_obj
print('Fungsi Objektif Populasi Update : \n',new_obj)
```

## Pengulangan Langkah CMBO

```python
iter = 1
while iter < max_iter:
  # Menghitung nilai fungsi Objektif
  F_obj = objective_function(agent)
            
  # Mengurutkan Populasi berdasarkan F_obj
  sorted_index = F_obj.argsort()
  F_obj = F_obj[sorted_index]
  agent_ori = agent.copy()
  agent = agent[sorted_index]

  # Membagi Populasi awal menjadi 2 Populasi (Tikus dan Kucing)
  upper_half = slice(gen_mice)
  lower_half = slice(gen_mice, len(agent))

  # Tikus
  mice_pop = agent[upper_half]
  mice_obj = F_obj[upper_half]

  # Kucing
  cat_pop = agent[lower_half]
  cat_obj = F_obj[lower_half]

  # Perubahan posisi Kucing
  cat_pop_move = update_cat_movement(cat_pop, mice_pop)
  cat_obj_move = objective_function(cat_pop_move)
  lower_than = cat_obj_move < cat_obj
  cat_obj[lower_than] = cat_obj_move[lower_than]
  cat_pop[lower_than] = cat_pop_move[lower_than]

  # Membentuk sarang
  haven_pop = generate_agent(lb, ub,size=(mice_pop.shape))
  haven_obj = objective_function(haven_pop)

  # Perubahan posisi tikus
  mice_pop_esc = update_mice_escape(mice_pop, haven_pop, mice_obj, haven_obj)
  mice_obj_esc = objective_function(mice_pop_esc)
            
  lower_than = mice_obj_esc < mice_obj
  mice_obj[lower_than] = mice_obj_esc[lower_than]
  mice_pop[lower_than] = mice_pop_esc[lower_than]
            
  # Agent Baru ( Gabungkan kedua perubahan antara populasi tikus dan kucing)
  new_obj = np.concatenate([mice_obj, cat_obj], axis=0)
  agent = np.concatenate([mice_pop, cat_pop], axis=0)
  
  iter+=1
  print()
  print('Iterasi ke-', iter)
  print('Populasi Tikus : \n', mice_pop)
  print('\n')
  print('Populasi Kucing: \n',cat_pop)
  print('\n')
  print('Fungsi Objektif: \n', new_obj)
```

## Hasil Akhir Metode CMBO

```python
# Solusi optimal
print('Solusi Optimal : \n', agent[0])
print('Fungsi Objektif: \n', new_obj[0])
```
